```python
# ----- TASK 1 -----
# Load data 
inputPath = "/share/students/bigdata/Dati/Lab2/OutputFolderLab1"
rawRDD = sc.textFile(inputPath)
```

```python
# 2. Keep lines starting with "ho" 
# Format is "word\tfreq" 
filteredRDD = rawRDD.filter(lambda line: line.split('\t')[0].startswith("ho"))
```

```python
# 3. Compute count and maximum frequency [cite: 14, 15]
countTask1 = filteredRDD.count()
# Convert freq part to integer for calculation [cite: 15]
freqsRDD = filteredRDD.map(lambda line: int(line.split('\t')[1]))
maxfreq = freqsRDD.max()
```

```python
# 4. Print Statistics for Task 1 [cite: 13]
print(f"Number of 'ho' lines: {countTask1}")
print(f"Maximum frequency (maxfreq): {maxfreq}")
```

```python
# ----- TASK 2 -----
# 1. Filter lines with freq > 80% of maxfreq
threshold = 0.8 * maxfreq
finalRDD = filteredRDD.filter(lambda line: int(line.split('\t')[1]) > threshold)
```

```python
# 2. Count final lines and print 
countTask2 = finalRDD.count()
print(f"Number of lines > 80% threshold: {countTask2}")
```

```python
# 3. Extract words only (remove frequency) 
wordsOnlyRDD = finalRDD.map(lambda line: line.split('\t')[0])
```

```python
# 4. Save results to your home directory [cite: 22, 26]
outputPath = "Lab5_Output"
wordsOnlyRDD.saveAsTextFile(outputPath)
```
